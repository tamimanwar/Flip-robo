```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the most viewed YouTube videos
table = soup.find("table", class_="wikitable")

# Loop through rows of the table (skip the header row)
for row in table.find_all("tr")[1:]:
    columns = row.find_all("td")
    
    rank = columns[0].get_text().strip()
    name = columns[1].get_text().strip()
    artist = columns[2].get_text().strip()
    upload_date = columns[4].get_text().strip()
    views = columns[5].get_text().strip()
    
    print("Rank:", rank)
    print("Name:", name)
    print("Artist:", artist)
    print("Upload Date:", upload_date)
    print("Views:", views)
    print("="*30)
```

```python
import requests
import requests
from bs4 import BeautifulSoup

# URL of the BCCI homepage
url = "https://www.bcci.tv/"

# Send a GET request to the homepage
response = requests.get(url)
homepage_content = response.content

# Parse the homepage using Beautiful Soup
homepage_soup = BeautifulSoup(homepage_content, "html.parser")

# Find the link to the international fixtures page
international_fixtures_link_elem = homepage_soup.find("a", text="International Fixtures")

if international_fixtures_link_elem:
    international_fixtures_link = international_fixtures_link_elem["href"]

    # Fetch the international fixtures page
    fixtures_url = "https://www.bcci.tv" + international_fixtures_link
    fixtures_response = requests.get(fixtures_url)
    fixtures_content = fixtures_response.content

    # Parse the international fixtures page using Beautiful Soup
    fixtures_soup = BeautifulSoup(fixtures_content, "html.parser")

    # Find and extract the required details (example)
    fixtures = fixtures_soup.find_all("div", class_="fixture__description")
    for fixture in fixtures:
        match_title = fixture.find("p", class_="fixture__format").get_text()
        series = fixture.find("span", class_="u-unskewed-text").get_text()
        place = fixture.find("p", class_="fixture__additional-info").get_text()
        date = fixture.find("span", class_="fixture__date").get_text()
        time = fixture.find("span", class_="fixture__time").get_text()

        print("Match Title:", match_title)
        print("Series:", series)
        print("Place:", place)
        print("Date:", date)
        print("Time:", time)
        print("=" * 30)
else:
    print("International Fixtures link not found.")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Statistics Times homepage
url = "http://statisticstimes.com/"

# Send a GET request to the homepage
response = requests.get(url)
homepage_content = response.content

# Parse the homepage using Beautiful Soup
homepage_soup = BeautifulSoup(homepage_content, "html.parser")

# Find the link to the Economy page
economy_link_elem = homepage_soup.find("a", text="Economy")

if economy_link_elem:
    economy_link = economy_link_elem["href"]

    # Fetch the Economy page
    economy_url = "http://statisticstimes.com" + economy_link
    economy_response = requests.get(economy_url)
    economy_content = economy_response.content

    # Parse the Economy page using Beautiful Soup
    economy_soup = BeautifulSoup(economy_content, "html.parser")

    # Find and extract the required details (example)
    table = economy_soup.find("table", class_="display dataTable")
    rows = table.find_all("tr")

    for row in rows[1:]:
        columns = row.find_all("td")
        
        rank = columns[0].get_text()
        state = columns[1].get_text()
        gdp_18_19 = columns[2].get_text()
        gdp_19_20 = columns[3].get_text()
        share_18_19 = columns[4].get_text()
        gdp_billion = columns[5].get_text()
        
        print("Rank:", rank)
        print("State:", state)
        print("GSDP(18-19):", gdp_18_19)
        print("GSDP(19-20):", gdp_19_20)
        print("Share(18-19):", share_18_19)
        print("GDP($ billion):", gdp_billion)
        print("=" * 30)
else:
    print("Economy link not found.")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the GitHub Trending page
url = "https://github.com/"

# Send a GET request to the Trending page
response = requests.get(url)
trending_content = response.content

# Parse the Trending page using Beautiful Soup
trending_soup = BeautifulSoup(trending_content, "html.parser")

# Find and extract the required details (example)
repos = trending_soup.find_all("article", class_="Box-row")
for repo in repos:
    title = repo.find("h1", class_="h3 lh-condensed").get_text().strip()
    description = repo.find("p", class_="col-9 color-text-secondary my-1 pr-4").get_text().strip()
    contributors_count = repo.find("a", class_="Link--muted d-inline-block mr-3").get_text().strip()
    language = repo.find("span", itemprop="programmingLanguage").get_text().strip()

    print("Repository Title:", title)
    print("Repository Description:", description)
    print("Contributors Count:", contributors_count)
    print("Language Used:", language)
    print("=" * 30)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the Billboard Hot 100 chart page
url = "https://www.billboard.com/charts/hot-100"

# Send a GET request to the Hot 100 chart page
response = requests.get(url)
chart_content = response.content

# Parse the chart page using Beautiful Soup
chart_soup = BeautifulSoup(chart_content, "html.parser")

# Find and extract the required details (example)
songs = chart_soup.find_all("li", class_="chart-list__element")
for song in songs:
    song_name = song.find("span", class_="chart-element__information__song").get_text()
    artist_name = song.find("span", class_="chart-element__information__artist").get_text()
    last_week_rank = song.find("span", class_="chart-element__meta text--last").get_text()
    peak_rank = song.find("span", class_="chart-element__meta text--peak").get_text()
    weeks_on_board = song.find("span", class_="chart-element__meta text--week").get_text()

    print("Song Name:", song_name)
    print("Artist Name:", artist_name)
    print("Last Week Rank:", last_week_rank)
    print("Peak Rank:", peak_rank)
    print("Weeks on Board:", weeks_on_board)
    print("=" * 30)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the highest selling novels page
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/"

# Send a GET request to the page
response = requests.get(url)
page_content = response.content

# Parse the page using Beautiful Soup
soup = BeautifulSoup(page_content, "html.parser")

# Find the table containing the highest selling novels
table = soup.find("table")

# Find and extract the required details
rows = table.find_all("tr")[1:]  # Skip the header row
for row in rows:
    columns = row.find_all("td")

    if len(columns) >= 6:  # Check if there are enough columns
        book_name = columns[0].get_text()
        author_name = columns[1].get_text()
        volumes_sold = columns[2].get_text()
        publisher = columns[3].get_text()
        genre = columns[4].get_text()

        print("Book Name:", book_name)
        print("Author Name:", author_name)
        print("Volumes Sold:", volumes_sold)
        print("Publisher:", publisher)
        print("Genre:", genre)
        print("=" * 30)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the most watched TV series page
url = "https://www.imdb.com/list/ls095964455/"

# Send a GET request to the page
response = requests.get(url)
page_content = response.content

# Parse the page using Beautiful Soup
soup = BeautifulSoup(page_content, "html.parser")

# Find the list of TV series
tv_series_list = soup.find_all("div", class_="lister-item-content")

# Loop through each TV series and extract the required details
for tv_series in tv_series_list:
    name = tv_series.h3.a.get_text()
    year_span = tv_series.h3.find("span", class_="lister-item-year").get_text()
    genre = tv_series.find("span", class_="genre").get_text().strip()
    run_time = tv_series.find("span", class_="runtime").get_text()
    ratings = tv_series.find("div", class_="ipl-rating-star").get_text().strip()
    votes = tv_series.find("span", attrs={"name": "nv"}).get("data-value")

    print("Name:", name)
    print("Year Span:", year_span)
    print("Genre:", genre)
    print("Run Time:", run_time)
    print("Ratings:", ratings)
    print("Votes:", votes)
    print("=" * 30)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the UCI Machine Learning Repository home page
url = "https://archive.ics.uci.edu/"

# Send a GET request to the home page
response = requests.get(url)
page_content = response.content

# Parse the home page using Beautiful Soup
soup = BeautifulSoup(page_content, "html.parser")

# Find the link to the "Show All Dataset" page using partial link text
show_all_link = soup.find("a", string="Show All Datasets")

# Check if the link was found
if show_all_link:
    show_all_datasets_url = url + show_all_link["href"]

    # Send a GET request to the "Show All Dataset" page
    datasets_response = requests.get(show_all_datasets_url)
    datasets_page_content = datasets_response.content

    # Parse the "Show All Dataset" page using Beautiful Soup
    datasets_soup = BeautifulSoup(datasets_page_content, "html.parser")

    # Find the table containing the dataset details
    datasets_table = datasets_soup.find("table", class_="datadisplay")

    # Find and extract the required details
    rows = datasets_table.find_all("tr")[1:]  # Skip the header row
    for row in rows:
        columns = row.find_all("td")

        dataset_name = columns[0].get_text()
        data_type = columns[1].get_text()
        task = columns[2].get_text()
        attribute_type = columns[3].get_text()
        num_instances = columns[4].get_text()
        num_attributes = columns[5].get_text()
        year = columns[6].get_text()

        print("Dataset Name:", dataset_name)
        print("Data Type:", data_type)
        print("Task:", task)
        print("Attribute Type:", attribute_type)
        print("Number of Instances:", num_instances)
        print("Number of Attributes:", num_attributes)
        print("Year:", year)
        print("=" * 30)
else:
    print("Link to 'Show All Datasets' page not found.")
```
