```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL of the webpage to scrape
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
```

```python
# Send a GET request to the URL and get the HTML content
response = requests.get(url)
```

```python
# Create a BeautifulSoup object with the response text and specify the parser
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Find all the article entries on the page
article_entries = soup.find_all('li', class_='pod-listing')
```

```python
# Lists to store the details of each article
titles = []
authors = []
published_dates = []
paper_urls = []
```

```python
# Loop through each article entry and extract the details
for entry in article_entries:
    # Extract the title
    title = entry.find('a', class_='pod-listing-header-link').text.strip()
    titles.append(title)

    # Extract the authors
    authors_list = entry.find('ol', class_='authorGroup')
    author_names = [author.text.strip() for author in authors_list.find_all('a')]
    authors.append(", ".join(author_names))

    # Extract the published date
    published_date = entry.find('span', class_='pod-listing-page-length-date').text.strip().split("|")[1].strip()
    published_dates.append(published_date)

    # Extract the paper URL
    paper_url = entry.find('a', class_='pod-listing-header-link')['href']
    paper_urls.append(paper_url)
```

```python
# Create a DataFrame with the scraped data
data = {
    'Paper Title': titles,
    'Authors': authors,
    'Published Date': published_dates,
    'Paper URL': paper_urls
}
```

```python
df = pd.DataFrame(data)
```

```python
# Print the DataFrame
print(df)
```
