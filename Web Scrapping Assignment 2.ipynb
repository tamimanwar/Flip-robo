```python
import time, os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Create a new instance of Chrome Options
options = webdriver.ChromeOptions()

# The "--start-maximized" argument tells Chrome to start in maximized mode, 
# i.e., fullscreen. This can be helpful when the visibility or clickability of 
# an element depends on the screen size.
options.add_argument("--start-maximized")

# Pass the options when initializing the webdriver instance. 
# This creates a new instance of the Chrome driver with the specified options.
driver = webdriver.Chrome(options=options)
```

# Question #1

```python
# Open the naukri.com website
driver.get("http://www.naukri.com/")

# Wait until the search input field is located
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_all_elements_located((By.XPATH, "//input[@class='suggestor-input ']")))

# Input the job title in the search field
driver.find_element(By.XPATH, "//input[@class='suggestor-input ']").send_keys("Data Analyst")

# Input the location in the search field
driver.find_element(By.XPATH, "//input[@placeholder='Enter location']").send_keys("Bangalore")

# Click the search button
driver.find_element(By.XPATH, "//div[@class='qsbSubmit']").click()

# Wait for the search results to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_all_elements_located((By.XPATH, "//article[@class='jobTuple']")))

# Initialize an empty list to store job data
job_rows = []

# Get the first 10 job postings
all_jobs = driver.find_elements(By.XPATH, "//article[@class='jobTuple']")[:10]
for job in all_jobs:
    # Extract job title, company name, experience required, and job location
    job_title = job.find_element(By.XPATH, ".//a[@class='title ellipsis']").text
    company_name = job.find_element(By.XPATH, ".//a[@class='subTitle ellipsis fleft']").text
    job_experience = job.find_element(By.XPATH, ".//li[@class='fleft br2 placeHolderLi experience']//span").text
    job_location = job.find_element(By.XPATH, ".//li[@class='fleft br2 placeHolderLi location']//span").text

    # Create a dictionary to store job details
    row = {
        "job_title": job_title,
        "job_location": job_location,
        "company_name": company_name,
        "experience_required": job_experience
    }

    # Append the dictionary to the list
    job_rows.append(row)

# Create a pandas DataFrame from the list of job data
df = pd.DataFrame(job_rows)

# Display the DataFrame
df
```

# Question #4

```python
# Open the flipkart.com website
driver.get("https://www.flipkart.com/")

# Wait until the search input field is located
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, "//input[@title='Search for Products, Brands and More']")))

# Input the product name in the search field
driver.find_element(By.XPATH, "//input[@title='Search for Products, Brands and More']").send_keys("sunglasses")

# Click the search button
driver.find_element(By.XPATH, "//button[@type='submit']").click()

# Initialize an empty list to store product data
sunglass_products = []

# Loop until 100 products have been collected
while True:
    # Wait for the product listings to load
    wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@style='transform: translate3d(0px, 0px, 0px);']")))

    # Get all the product listings on the page
    all_products = driver.find_elements(By.XPATH, "//div[@style='transform: translate3d(0px, 0px, 0px);']")

    # Extract data from each product listing
    for product in all_products:
        brand = product.find_element(By.XPATH, "./div[1]").text
        product_desc = product.find_element(By.XPATH, "./a[1]").text
        price = product.find_element(By.XPATH, "./a[2]/div/div").text

        # Create a dictionary to store product details
        product_dict = {
            "brand": brand,
            "product-description": product_desc,
            "price": price
        }

        # Append the dictionary to the list
        sunglass_products.append(product_dict)

    # Stop collecting data after 100 products have been collected
    if len(sunglass_products) >= 100:
        sunglass_products = sunglass_products[:100]
        break

    # Go to the next page of search results
    driver.find_element(By.XPATH, "//a[contains(., 'Next')]").click()

    # Pause execution for 3 seconds to allow the page to load
    time.sleep(3)

# Create a pandas DataFrame from the list of product data
df = pd.DataFrame(sunglass_products)

# Display the DataFrame
df
```

```python
# Open the product review page for the Apple iPhone 11
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market%20place=FLIPKART")

# Wait until the reviews are located
wait = WebDriverWait(driver, 10)

# Initialize an empty list to store review data
reviews = []

# Loop until 100 reviews have been collected
while True:
    # Wait for the reviews to load
    wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='col']")))

    # Get all the reviews on the page, excluding the first one
    all_reviews = driver.find_elements(By.XPATH, "//div[@class='col']")[1:]

    # Extract data from each review
    for review in all_reviews:
        rating = review.find_element(By.XPATH, ".//div[@class='row'][1]/div").text
        review_summary = review.find_element(By.XPATH, ".//div[@class='row'][1]/p").text
        full_review = review.find_element(By.XPATH, ".//div[@class='row'][2]//div").text

        # Create a dictionary to store review details
        product_dict = {
            "rating": rating,
            "review-summary": review_summary,
            "full-review": full_review
        }

        # Append the dictionary to the list
        reviews.append(product_dict)

    # Stop collecting data after 100 reviews have been collected
    if len(reviews) >= 100:
        reviews = reviews[:100]
        break

    # Go to the next page of reviews
    driver.find_element(By.XPATH, "//a[contains(., 'Next')]").click()

    # Pause execution for 3 seconds to allow the page to load
    time.sleep(3)

# Create a pandas DataFrame from the list of review data
df = pd.DataFrame(reviews)

# Display the DataFrame
df
```

# Question #8

```python
# Open the azquotes.com website
driver.get("https://www.azquotes.com/")

# Wait until the 'Top Quotes' link is located
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, "//a[contains(., 'Top Quotes')]")))

# Click on the 'Top Quotes' link
driver.find_element(By.XPATH, "//a[contains(., 'Top Quotes')]").click()

# Initialize an empty list to store quotes data
thousand_quotes = []

# Loop until 1000 quotes have been collected
while True:
    # Wait until the quote listings are located
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_all_elements_located((By.XPATH, "//ul[@class='list-quotes']/li")))

    # Get all the quotes on the page
    quotes = driver.find_elements(By.XPATH, "//ul[@class='list-quotes']/li")
    for quote in quotes:
        # Get the quote text
        quote_text = quote.find_element(By.XPATH, ".//a[@class='title']").text
        # Get the author name
        author_name = quote.find_element(By.XPATH, ".//div[@class='author']").text
        # Get the tags
        tags = ""
        all_tags = quote.find_elements(By.XPATH, ".//div[@class='tags']/a")
        for tag in all_tags:
            tags += tag.text + ", "

        # Create a dictionary to store quote details
        quote_dict = {
            "quote": quote_text,
            "author": author_name,
            "tags": tags
        }

        # Append the dictionary to the list
        thousand_quotes.append(quote_dict)

    # Stop collecting data after 1000 quotes have been collected
    if len(thousand_quotes) >= 1000:
        thousand_quotes = thousand_quotes[:1000]
        break

    # Go to the next page of quotes
    try:
        driver.find_element(By.XPATH, "//a[contains(., 'Next')]").click()
    except:
        break

# Create a pandas DataFrame from the list of quotes data
df = pd.DataFrame(thousand_quotes)

# Display the DataFrame
df
```

# Question #9

```python
# Open the jagranjosh.com website
driver.get("http://www.jagranjosh.com/")

# Wait until the 'GK' link is located
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, "//a[contains(., 'GK')]")))

# Click on the 'GK' link
driver.find_element(By.XPATH, "//a[contains(., 'GK')]").click()

# Pause execution for 3 seconds to allow the page to load
time.sleep(3)

# Click on the 'List of all Prime Ministers of India' link
driver.find_element(By.XPATH, "//a[contains(., 'List of all Prime Ministers of India')]").send_keys(Keys.ENTER)

# Wait until the table of prime ministers is located
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, "//table/tbody[1]/tr")))

# Initialize an empty list to store prime minister data
pms_list = []

# Get all the rows in the table
pm_table_rows = driver.find_elements(By.XPATH, "//table/tbody[1]/tr")

# Loop through each row in the table
for row in pm_table_rows:
    try:
        # Get the prime minister's name
        name = row.find_element(By.XPATH, "./td[2]").text
        # Get the prime minister's birth and death years
        born_dead = row.find_element(By.XPATH, "./td[3]").text
        # Get the prime minister's term of office
        term_of_office = row.find_element(By.XPATH, "./td[4]").text
        # Get any remarks about the prime minister
        remark = row.find_element(By.XPATH, "./td[5]").text
    except:
        continue

    # Create a dictionary to store the prime minister's details
    pm_dict = {
        "name": name,
        "born-dead": born_dead,
        "term-of-office": term_of_office,
        "remark": remark
    }

    # Append the dictionary to the list
    pms_list.append(pm_dict)

# Create a pandas DataFrame from the list of prime minister data
df = pd.DataFrame(pms_list)

# Display the DataFrame
df
```

```python
# Open the motor1.com website
driver.get("https://www.motor1.com/")

# Wait until the search bar is located
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.ID, "search_input")))

# Find the search bar and enter "50 most expensive cars"
search = driver.find_element(By.ID, "search_input")
search.send_keys("50 most expensive cars")
search.send_keys(Keys.ENTER)

# Pause execution for 3 seconds to allow the search results to load
time.sleep(3)

# Click on the link to the '50 Most Expensive Cars In The World' article
driver.find_element(By.XPATH, "//a[contains(., '50 Most Expensive Cars In The World')]").click()

# Initialize an empty list to store car data
expensive_cars = []

# Get all the cars in the list
all_cars = driver.find_elements(By.XPATH, "//ul")[1].find_elements(By.XPATH, "./li")

# Loop through each car in the list
for car in all_cars:
    try:
        # Get the car's name and price, which are separated by " - "
        car_name = car.text.split(" - ")[0]
        car_price = car.text.split(" - ")[1]
    except:
        # If the car's name and price are not separated by " - ", try splitting by ":"
        car_name = car.text.split(":")[0]
        car_price = car.text.split(":")[1]

    # Create a dictionary to store the car's details
    car_dict = {
        "car-name": car_name,
        "price": car_price
    }

    # Append the dictionary to the list
    expensive_cars.append(car_dict)

# Create a pandas DataFrame from the list of car data
df = pd.DataFrame(expensive_cars)

# Display the DataFrame
df
```
